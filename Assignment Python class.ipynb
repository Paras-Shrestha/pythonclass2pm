```python
# WAP to create a (2*2) matrix using LIST inside LIST.
matrix=[]
r=int(input("enter the no of row="))
c=int(input("enter the name of the column="))
for i  in range(r):
    a=[]
    for j in range(c):
        b=int(input("enter the no="))
        a.append(b)
    matrix.append(a)

for k in matrix:
        for l in k:
            print(l, end=" ")
        print()
```
