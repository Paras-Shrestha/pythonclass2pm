```python
from bs4 import BeautifulSoup
import requests
soup = []
for i in range(1,11):
    url=BeautifulSoup(f"http://www.nepalstock.com/main/todays_price/index/{i}/",'html.parser')
    soup.append(requests.get(url))
```

```python
tag = []
print(len(soup))
for i in range(0,10):
  lxml_code=BeautifulSoup(soup[i].text,"lxml")
  table = lxml_code.table
  tag.append(table.find_all('tr'))
```

```python
print(len(tag))
output = []
for i in range(0,10):
  innerTag = tag[i]
  startIndex = 1
  if(i != 0):
    startIndex = 2
  for j in range(startIndex,22):
    output.append(innerTag[j].text.split('\n')[1:-1])
output
```

```python
import csv 
file = open('nepsetest.csv','w')
x=csv.writer(file)
for i in output:
    x.writerow(i)
```
