```python
from bs4 import BeautifulSoup
import requests
url=BeautifulSoup('http://www.nepalstock.com/todaysprice','html.parser')
soup=requests.get(url)
soup
```

```python
code=soup.text
code
```

```python
lxml_code=BeautifulSoup(code,"lxml")
lxml_code
```

```python
lxml_code.table
```

```python
table=lxml_code.table
```

```python
tag=table.find_all('tr')
tag
```

```python
table.find('index 1')
```

```python
tag[1].text.split("\n")
```

```python
for i in range(1,22):
    print(tag[i].text.split('\n')[1:-1])
    
```

```python
output=[]
for i in range(1,22):
    datas=tag[i].text.split('\n')[1:-1]
    output.append(datas)
        
output
```

```python
import csv 
file = open('nepse2.csv','w')
x=csv.writer(file)
for i in output:
    x.writerow(i)
```
